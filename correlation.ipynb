```python
import pandas as pd
import numpy as np
file = "D:\codes_python\sds project\games_normalized.csv"
```

```python
games = pd.read_csv(file)
```

```python
games.drop(["Unnamed: 0"], axis = 1, inplace = True)
```

```python
games
```

```python
games.corr(method = "pearson")
```

```python
games.corr(method = "kendall")
```

```python
games.corr(method = "spearman")
```

```python
from itertools import combinations
columns = ["rated", "turns", "white_rating","black_rating", "opening_ply", "game_time"]
columns_combinations = list(combinations(columns, 2))
```

```python
columns_combinations
```

```python
import matplotlib.pyplot as plt
for i in columns_combinations:
    plt.scatter(games[i[0]], games[i[1]])
    plt.xlabel(i[0]) 
    plt.ylabel(i[1]) 
    plt.show()
```
