We will prove/disprove the following using hypothesis testing:<br>
1) Is the mean white rating greater than the mean black rating at 95% level ?<br>
2) Is the winner of games most of the time "white" at a 90% level ?<br>
3) Is the most common opening in all the games ["e4", "e5"] at a 95% level ?<br>
4) Do most of the games end in a checkmate at a 85% level ?<br>
5) Do most of the games end in the range of 40 to 70 moves at a 99% level ?<br>
6) Are more than 50% of the games rated at a 90% level ?<br>

```python
from scipy.stats import norm

# p_values = scipy.stats.norm.sf(abs(z_scores)) #one-sided
# p_values = scipy.stats.norm.sf(abs(z_scores))*2 #twosided
```

```python
#Is the mean white rating greater than the mean black rating at 95% level?
alpha = 0.05
xbar = games["white_rating"].mean()
ybar = games["black_rating"].mean()
sigma_square_x = games["white_rating"].var()
sigma_square_y = games["black_rating"].var()
delta0 = 0
z = (xbar - ybar - delta0)/((sigma_square_x/len(games["white_rating"])) + 
                   (sigma_square_y/len(games["black_rating"])))**0.5
# Here the H0 is xbar - ybar <= 0
# H1 is xbar - ybar > 0, thus we need to take the area to the right of z
p_value = 1 - norm.cdf(abs(z)) 

if p_value <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

So in the above claim we fail to reject H0. 
Let us try at 90% level

```python
alpha = 0.1
xbar = games["white_rating"].mean()
ybar = games["black_rating"].mean()
sigma_square_x = games["white_rating"].var()
sigma_square_y = games["black_rating"].var()
delta0 = 0
z = (xbar - ybar - delta0)/((sigma_square_x/len(games["white_rating"])) + 
                   (sigma_square_y/len(games["black_rating"])))**0.5
# Here the H0 is xbar - ybar <= 0
# H1 is xbar - ybar > 0, thus we need to take the area to the right of z
p_value = 1 - norm.cdf(z) 

if p_value <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

So in the above claim we rejected H0. So we can say that the mean white rating is greater than the mean black rating at a 90% level.

```python
#Is the winner of games most of the time "white" at a 90% level ?
alpha = 0.1
phat = games["winner"].value_counts(1.0)["white"]
p0 = 0.3750 #data taken from "https://en.wikipedia.org/wiki/First-move_advantage_in_chess#Winning_percentages"
n = len(games["winner"])

#Here H0 is that white is not the winner in most of the games that is p0 <= 0.3750
#Thus H1 would be that white is the winner in most of the games that is p0 > 0.3750

z = (phat - p0)/((p0*(1 - p0))/n)**0.5
p_value = 1 - norm.cdf(z) 

if p_value <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

As per the above claim we reject H0 at 90% level, that is white is not the winner in most of the games.

```python
#Is the most common opening in all the games not ["e4", "e5"] anymore at a 95% level ?
alpha = 0.05
e4e5 = 0
for i in games["moves"]:
    if i[2] + i[3] == "e4" and i[8] + i[9] == "e5":
        e4e5 += 1
phat = e4e5 / len(games["moves"])
p0 = 0.5

#Here H0 is that most games begin with the opening e4, e5 that is p0 >= 0.5
#Thus H1 would be that most games donot begin with the opening e4, e5 that is p0 < 0.5
 
z = (phat - p0)/((p0*(1 - p0))/n)**0.5
p_value = norm.cdf(z) 

if p_value <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

As per the claim above we reject H0, that is most of the games begin with the opening e4, e5, at a 95% level.

```python
#Do most of the games not end in a checkmate at a 85% level ?
alpha = 0.15
phat = games["victory_status"].value_counts(1.0)["mate"]
p0 = 0.25

#Here H0 is that most games end in a checkmate that is p0 >= 0.25
#Thus H1 would be that most games donot end in a checkmate that is p0 < 0.25
 
z = (phat - p0)/((p0*(1 - p0))/n)**0.5
p_value = norm.cdf(z) 

if p_value <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

According to the above claim, we fail to reject H0 at a 85% level, that is most of the games do end in checkmate.

```python
#Do most of the games end in the range of 40 to 70 moves at a 99% level ?
alpha = 0.01
phat = sum(games["turns"].value_counts()[i] for i in set(games["turns"]) if 40 <= i <= 70) / len(games["turns"])
p0 = 0.70

#Here H0 is that most games do not end in the range of 40 to 70 moves that is p0 <= 0.70
#Thus H1 would be that most games do end in the range of 40 to 70 moves that is p0 > 0.70
 
z = (phat - p0)/((p0*(1 - p0))/n)**0.5
p_value = 1 - norm.cdf(z) 

if p_value <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

As we can see the claim made above, we fail to reject the null hypothesis H0 that is most of the games donot end in the range of 40 to 70 moves

```python
#Are more than 50% of the games rated at a 90% level ?
alpha = 0.1
phat = games["rated"].mean()
p0 = 0.50

#Here H0 is that most games are not rated that is p0 <= 0.50
#Thus H1 would be that most games are rated that is p0 > 0.50
 
z = (phat - p0)/((p0*(1 - p0))/n)**0.5
p_value = 1 - norm.cdf(z) 

if p_value <= alpha:
    print("Reject H0")
else:
    print("Fail to reject H0")
```

As we can see the above result, we reject H0 at a 90% level, that is more that 50% of the games are not rated.
