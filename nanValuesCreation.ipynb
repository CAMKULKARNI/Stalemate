```python
columns = list(new_games.columns)
```

```python
columns
```

```python
#creating null values within the data frame
import random
for i in columns:
    print(new_games[i].iloc[0], "->", type(new_games[i].iloc[0]) , "\n")
```

```python
import numpy
for _ in range(450):
    column = random.choice(columns)
    row = random.randrange(0, 9367)
    if type(new_games[column].iloc[row]) == numpy.int64:
        new_games[column].iloc[row] = numpy.nan
    elif type(new_games[column].iloc[row]) == str:
        new_games[column].iloc[row] = None
    elif type(new_games[column].iloc[row]) == numpy.float64:
        new_games[column].iloc[row] = numpy.nan
    elif type(new_games[column].iloc[row]) == numpy.bool_:
         new_games[column].iloc[row] = None
```

```python
new_games
```

```python
total = 0
for i in new_games:
    total += new_games[i].isnull().sum()
    print(i, "->", new_games[i].isnull().sum())
print("total null values : ", total)
```
